```python

#1
def chatbot():
    cars = {"Ertiga":50, "Swift":45, "Innova":80}

    print("Available Cars:")
    for car in cars:
        print(car)

    name = input("Enter name: ")   # fixed
    choice = input("Choose car: ")

    if choice not in cars:
        print("Car not available")
        return

    days = int(input("Enter days: "))
    cost = cars[choice] * days

    print("\nBooking Confirmed!")
    print("Name:", name)
    print("Car:", choice)
    print("Total Cost:", cost)

chatbot()
```

```python
#4

import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.cluster import AgglomerativeClustering
from sklearn.metrics import silhouette_score

# Load dataset
data = pd.read_csv("/Mall_Customers.xls")

# Select Annual Income and Spending Score
X = data.iloc[:, [3,4]].values

# Train-Test Split
X_train, X_test = train_test_split(X, test_size=0.2)

# Apply Agglomerative Clustering
model = AgglomerativeClustering(n_clusters=5)
y = model.fit_predict(X_train)

# Evaluate model
score = silhouette_score(X_train, y)
print("Silhouette Score:", score)

# Plot clusters
plt.scatter(X_train[:,0], X_train[:,1], c=y)
plt.xlabel("Annual Income")
plt.ylabel("Spending Score")
plt.title("Customer Segmentation")
plt.show()
```

```python

#6
import pandas as pd
from apyori import apriori

# Load dataset
data = pd.read_csv("/Market_Basket_Optimisation.csv", header=None)

# Convert dataset into list of transactions
transactions = []
for i in range(7501):
    transactions.append([str(data.values[i,j]) for j in range(20)])

# Apply Apriori Algorithm
rules = apriori(transactions,
                min_support=0.003,
                min_confidence=0.2,
                min_lift=3,
                min_length=2)

# Display association rules
results = list(rules)

for item in results:
    pair = item[0]
    items = [x for x in pair]

    print("Rule:", items)
    print("Support:", item[1])
    print("Confidence:", item[2][0][2])
    print("Lift:", item[2][0][3])
    print("---------------")
```

```python
from queue import Queue

# Graph
graph = {
    '5': ['3', '7'],
    '3': ['2', '4'],
    '7': ['8'],
    '2': [],
    '4': [],
    '8': ['10', '20'],
    '10': [],
    '20': []
}

def bfs(root):
    visited = set()
    q = Queue()

    visited.add(root)
    q.put(root)

    while not q.empty():
        node = q.get()
        print(node, end=" ")

        for n in graph[node]:
            if n not in visited:
                visited.add(n)
                q.put(n)

# Function call
print("BFS Traversal:")
bfs('5')
```

```python
# Graph
graph = {
    '5': ['3', '7'],
    '3': ['2', '4'],
    '7': ['8'],
    '2': [],
    '4': [],
    '8': ['10', '20'],
    '10': [],
    '20': []
}

def dfs(node, visited):
    visited.add(node)
    print(node, end=" ")

    for n in graph[node]:
        if n not in visited:
            dfs(n, visited)

# Function call
print("DFS Traversal:")
visited = set()
dfs('5', visited)
```
